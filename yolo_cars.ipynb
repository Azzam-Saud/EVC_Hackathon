```python
import cv2
from ultralytics import YOLO
```

```python
model = YOLO('best.pt') 

model.conf = 0.25

video_path = 'D:/Azzam/Personal_Projects/EVC/Python/EVC_Hackathon/video2.mp4'
output_path = 'D:/Azzam/Personal_Projects/EVC/Python/EVC_Hackathon/output_video.mp4'

cap = cv2.VideoCapture(video_path)
```

```python
frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = cap.get(cv2.CAP_PROP_FPS)

fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter(output_path, fourcc, fps, (frame_width, frame_height))
```

```python
image_path = 'D:\Azzam\Personal_Projects\EVC\Python\EVC_Hackathon\img9.jpg'

small_image = cv2.imread(image_path)
resize_scale = 0.5 
small_image = cv2.resize(small_image, (0, 0), fx=resize_scale, fy=resize_scale)
small_image_height, small_image_width = small_image.shape[:2]

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    results = model(frame)
    
    boxes = results[0].boxes.xyxy.cpu().numpy()  # Bounding boxes
    scores = results[0].boxes.conf.cpu().numpy()  # Confidence scores
    class_ids = results[0].boxes.cls.cpu().numpy()  # Class IDs
    
    class_names = model.names  
    
    # Count detected vehicles
    vehicle_count = 0
    
    # Annotate the frame
    for box, score, class_id in zip(boxes, scores, class_ids):
        x1, y1, x2, y2 = map(int, box)
        if class_names[int(class_id)] == "car":  
            label = f'Vehicle {score:.2f}'
            cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
            cv2.putText(frame, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)
            vehicle_count += 1

    cv2.putText(frame, f'Vehicle Count: {vehicle_count}', (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
    
    current_time = cap.get(cv2.CAP_PROP_POS_MSEC) / 1000.0

    if 9 <= current_time <= 22:
        y_offset = 10
        x_offset = frame_width - small_image_width - 10
        frame[y_offset:y_offset + small_image_height, x_offset:x_offset + small_image_width] = small_image

    # Write the frame to the output video
    out.write(frame)

# Release everything
cap.release()
out.release()
```
